```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# droped unnamed column
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)

df = df.head()
print(df)
```

```python
#change weelky study hour column
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('  5 - 10', '', regex=".")
df.head()
```

```python
plt.figure(figsize=(8, 6))
ax=sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df=pd.read_csv('/content/Expanded_data_with_more_features.csv')
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean',"ReadingScore":'mean','WritingScore':'mean'})
sns.heatmap(gb, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title('Average Scores by Parent Education')
plt.show()
```

Fom th eabove chrt we have concluded that the education of the parentshava a good impact on their scores

```python
gb2=df.groupby('ParentMaritalStatus').agg({'MathScore': 'mean',"ReadingScore":'mean','WritingScore':'mean'})
sns.heatmap(gb2, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title('Average Scores by Parent Marital Status')
plt.show()
```
